Median = 12

 Most representative value: Median (12 dozens)

Why?
- The sales are fairly balanced but include slightly higher values (14, 15). The median best represents a “typical” day without being influenced by these higher values.

4. A car dealer notes the prices of used cars: [$8,000, $9,500, $10,200, $11,000, $50,000].

Why is the median a better measure than the mean in this case? Calculate the median.

- Data: [$8,000, $9,500, $10,200, $11,000, $50,000]
Sorted: 8,000 | 9,500 | 10,200 | 11,000 | 50,000

Median = $10,200

Why median is better than mean?

- The extreme price ($50,000) skews the mean upward. The median reflects the typical used-car price more accurately.

5. A student times how long it takes to finish a puzzle each day: [25, 30, 27, 35, 40].

- Data: [25, 30, 27, 35, 40]
Range = 40 − 25 = 15 minutes

Interpretation:
The student’s puzzle-solving time varies by 15 minutes, showing noticeable inconsistency from day to day.
